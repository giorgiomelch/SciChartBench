```python
from deplot import load_model_and_processor, process_images
from pathlib import Path

path_to_pmc_data = Path("../data/pmc")
output_path = Path("../predictions/DePlot/pmc")
model, processor, device = load_model_and_processor()
for chart_class in path_to_pmc_data.iterdir():
    for difficulty in chart_class.iterdir():
        print(f"\nElaborazione cartella: {chart_class.name}/{difficulty.name}.")
        cc_output_path = output_path / chart_class.name / difficulty.name
        process_images(list(difficulty.iterdir()), cc_output_path, model, processor, device)
```
